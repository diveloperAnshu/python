```python
import matplotlib.pyplot as plt
import numpy as np
```

USEFULFOR MAKING PLOTS

```python
x=np.linspace(0,10,100) # here we will get 100  equally spaced values between 0 and 10
y=np.sin(x) # it will find the sin values for particular x value
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
# cos wave
plt.plot(x,z)
plt.show()
```

```python
# adding title to the graphs (x-axis and y-axis labels)
plt.plot(x,y)
plt.xlabel("x value")
plt.ylabel("sine value")
plt.title("Sine Wave")
plt.show()
```
