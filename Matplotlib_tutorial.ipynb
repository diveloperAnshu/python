```python
import matplotlib.pyplot as plt
import numpy as np
```

USEFULFOR MAKING PLOTS

```python
x=np.linspace(0,10,100) # here we will get 100  equally spaced values between 0 and 10
y=np.sin(x) # it will find the sin values for particular x value
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
# cos wave
plt.plot(x,z)
plt.show()
```

```python
# adding title to the graphs (x-axis and y-axis labels)
plt.plot(x,y)
plt.xlabel("x value")
plt.ylabel("sine value")
plt.title("Sine Wave")
plt.show()
```

```python
#  more than 1 line in a single plot
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'r*')
plt.plot(x,np.cos(x),'g-')
plt.show()
```

BAR PLOT

```python
fig=plt.figure()
axis=fig.add_axes([0,0,1,1])
languages=['Odia','Hindi','English','Telugu','Tamil']
people=[200,500,700,250,300]
axis.set_xlabel('Languages')
axis.set_ylabel('No of people')
axis.set_title('No of people vs Languages')
axis.bar(languages,people)
plt.show()
```
