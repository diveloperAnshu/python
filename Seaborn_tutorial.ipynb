```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

DATA VISUALISATION LIBRARY

```python
#total bill vs tip dataset
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme()
```

```python
#visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
#taking he iris dataset

iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.relplot(data=iris,x='sepal_length',y='sepal_width',col='species',hue='species',style='species',size='petal_length')
```

```python
#here we can use scatterplot to visualize too
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
```

```python
#loading titanic dataset
titanic=sns.load_dataset('titanic')
titanic.head()
```

COUNT PLOT

```python
sns.countplot(x='class',data=titanic,hue='class')
sns.countplot(x='alive',data=titanic,hue='alive')
```

```python
sns.barplot(x='sex',y='survived',data=titanic,hue='class')
```
